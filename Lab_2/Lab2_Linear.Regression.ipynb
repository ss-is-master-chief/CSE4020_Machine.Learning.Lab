```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

## Simple Linear Regression

```python
# loading the dataset
df = pd.read_csv("headbrain1.csv")
df.head()
```

```python
X = df.iloc[:, 0:1]
Y = df.iloc[:, 1:2]

print("Shape of Independent Variable Vector, X: ", X.shape)
print("Shape of Dependent Variable Vector, Y:   ", Y.shape)
```

```python
# fitting (X, Y) to the Linear Regression Model
oneV_linear_reg = LinearRegression()
oneV_linear_reg.fit(X, Y)

# Predicting Y using the trained/fitted linear model
predicted_y = oneV_linear_reg.predict(X)
```

```python
plt.figure(figsize=(12,8))
# Plotting Scatter plot - X vs Y
plt.scatter(X, Y, color="green")
# Plotting Line plot - X vs predicted_Y
plt.plot(X, predicted_y, color="red")
plt.xlabel("Head Size(cm^3)")
plt.ylabel("Brain Weight(grams)")
plt.title("Simple Linear Regression")
plt.legend(["Prediction", "Source Data"])

print("Coefficient (Slope, m): ", oneV_linear_reg.coef_[0][0])
print("Intercept (c):\t\t", oneV_linear_reg.intercept_[0])
print("Equation: \t\t y = {0} * x + {1}".format(oneV_linear_reg.coef_[0][0], 
                                                oneV_linear_reg.intercept_[0]))
print("RMSE Score:\t\t", mean_squared_error(X, oneV_linear_reg.predict(X)))
```
